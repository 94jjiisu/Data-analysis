```python
lst = list(case.columns.values)

for i in lst:
    a = i
    if type(case[a][0]) == str:
        for j in range(len(case[a])):
            case[a][j] = case[a][j].replace(',', '')
```

```python
for i in lst:
    a = i
    if type(case[a][0]) == str:
        for j in range(len(case[a])):
            case[a][j] = case[a][j].replace('M', '')
```

```python
for i in lst:
    a = i
    if type(case[a][0]) == str:
        for j in range(len(case[a])):
            case[a][j] = case[a][j].replace('B', '')
```

```python
lst = list(case.columns.values)

for i in lst:
    a = i
    if type(case[a][0]) == str:
        for j in range(len(case[a])):
            case[a][j] = case[a][j].replace('%', '')
```

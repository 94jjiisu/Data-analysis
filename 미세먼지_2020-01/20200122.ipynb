데이터의 끝부분 자르기

```python
a = df.iloc[0:362]
```

변수 이름 다시 설정하기

```python
a.columns = 'Kdate', 'K1', 'tem','dir','dir1','vel','hum','coal','diesel','Chdate','Ch1','Ch2'
a.head()
```

index 바꾸기

```python
a = a.set_index('Kdate')
a.head()
```

필요없는 열 삭제

```python
del a['dir1']
del a['Chdate']
a.head()
```

결측된 데이터 찾기

```python
a.isnull().sum()
```

```python
a
```

결측된 데이터값 행 삭제

```python
b = a.dropna(axis=0)
b.describe()
```

그래프 그리기

```python
plt.plot(b['K1'])
```

```python
plt.plot(b['Ch1'])
plt.plot(b['Ch2'])
```

```python
plt.figure(figsize=(16,10))
plt.plot(b['K1'], label='Kor')
plt.plot(b['Ch1'], label ='Ch1')
plt.plot(b['Ch2'], label ='Ch2')
plt.legend()
plt.xlabel('date')
plt.ylabel('concentration')
plt.title('dust')
plt.show()
```

```python
c = b[['K1', 'Ch1','Ch2']]
c.plot()
```

```python
del b['coal']
del b['diesel']
b.head()
```

correlation 보기

```python
b.corr()
```

ols 검사하기

```python
import statsmodels.api as sm
model_sm = sm.OLS.from_formula("K1 ~ Ch1", data = b)
result_sm = model_sm.fit()
print(result_sm.summary())
```

```python
import statsmodels.api as sm
model_sm = sm.OLS.from_formula("K1 ~ Ch2", data = b)
result_sm = model_sm.fit()
print(result_sm.summary())
```

```python
model_sm = sm.OLS.from_formula("K1 ~ tem + vel + hum + Ch1 + Ch2", data = b)
result_sm = model_sm.fit()
print(result_sm.summary())
```

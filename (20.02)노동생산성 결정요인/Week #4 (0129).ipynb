```python
from linearmodels import PanelOLS
from linearmodels import PooledOLS
```

```python
data=data1
year=pd.Categorical(data.year)
data=data.set_index(['Country', 'year'])
data['year']=year
```

```python
mod = PanelOLS.from_formula('product_emp ~ work_hour + work_hour_sq + unemp + gap + RnD + min_wage + min_wage_sq + edu_tertiary', data)
result1_re = mod.fit(cov_type = 'clustered', cluster_entity=True)
print(result1_re)
```

```python
mod = PooledOLS.from_formula('product_emp ~ work_hour + work_hour_sq + unemp + gap + RnD + min_wage + min_wage_sq + edu_tertiary', data)
result1_ols = mod.fit()
print(result1_ols)
```

```python
mod = PanelOLS.from_formula('product_emp ~ work_hour + work_hour_sq + gap + RnD + min_wage + min_wage_sq + EntityEffects', data)
result1_fe = mod.fit(cov_type = 'clustered', cluster_entity=True)
print(result1_fe)
```

```python
df = pd.read_csv("data2.csv")
df
```

```python
data=df
year=pd.Categorical(data.year)
data=data.set_index(['Country', 'year'])
data['year']=year
```

```python
mod = PanelOLS.from_formula('product_emp ~ work_hour + work_hour_sq + unemp + gap + RnD + min_wage + min_wage_sq + EntityEffects', data)
result2_fe = mod.fit(cov_type = 'clustered', cluster_entity=True)
print(result2_fe)
```

```python
from collections import OrderedDict
from linearmodels.panel import compare
```

```python
res = OrderedDict()
res['OLS'] = result1_ols
res['Random'] = result1_re
res['Fixed'] = result1_fe
print(compare(res))
```
